# Cleaning Data

```python
# convert a date into numerical value
games['date'] = pd.to_datetime(games['date'])
```

```python
# convert "venue" coulmn into numerical value
# convert from string to categories then to numbers
games['venue_num'] = games['venue'].astype('category').cat.codes
```

```python
games['opponent_num'] = games['opponent'].astype('category').cat.codes
```

```python
# ading time_hour column which get rid of mins then convert it into intger
games['time_hour'] = games['time'].str.replace(':.+', '', regex=True).astype('int')
```

```python
games['day_num'] = games['date'].dt.day_of_week
```

```python
games['target'] = (games['result'] == 'W').astype('int')
```

### improving precision

grouped_games = games.groupby

```python
grouped_games = games.groupby('team')
```

```python
group = grouped_games.get_group('Arsenal')
```

```python
group
```

```python
def rol_averages(group, cols, new_cols):
    group = group.sort_values('date')
    rol_stats = group[cols].rolling(3, closed='left').mean() # take the current week out
    group[new_cols] = rol_stats
    group = group.dropna(subset=new_cols) # drop any missing value
    return group
```

```python
cols = ['gf', 'ga', 'sh', 'sot', 'dist', 'fk', 'pk','pkatt']
new_cols = [f'{c}_rolling' for c in cols]
```

```python
new_cols
```

```python
rol_averages(group, cols, new_cols)
```

```python
games_rolling = games.groupby('team').apply(lambda x: rol_averages(x, cols, new_cols)) # applying a function to each team
```

```python
games_rolling
```

```python
games_rolling = games_rolling.droplevel('team') # drop index level (team)
```

```python
games_rolling
```

```python
games_rolling.index = range(games_rolling.shape[0]) # assign new indices to have uniqe value for each index
```

```python
games_rolling
```

### Retrain The Model

```python
def make_predictions(data, predictors):
    train = data[data['date'] < '2022-01-01']
    test = data[data['date'] > '2022-01-01']
    rfc.fit(train[predictors], train['target'])
    preds = rfc.predict(test[predictors])
    combined = pd.DataFrame(dict(actual=test['target'], predicted = preds), index=test.index)
    precision = precision_score(test['target'], preds)
    return combined, precision
```

```python
combined, precision = make_predictions(games_rolling, predictors + new_cols)
```
